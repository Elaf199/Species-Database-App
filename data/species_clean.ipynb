```python
import pandas as pd
```

```python
file=pd.read_excel("new_data.xlsx")
file.head()
```

```python
file.info()
print(file.columns)
```

```python
file=file.dropna(how='all')
file=file.reset_index(drop=True)
for col in file.columns:
    if file[col].dtype == "object":
        file[col] = file[col].str.strip()
```

```python
# fixing the column Sr No
file["Sr No"]= file['Sr No'].astype(int)
file.head()
```

```python
#missing values and duplicates checking
file = file.fillna("")
file = file.drop_duplicates(subset=["Scientific name"])
```

```python
file.to_excel("cleaned_data.xlsx", index=False)
```

```python
print(file.columns)
```

```python
import matplotlib.pyplot as plt

leaf_counts = file['Leaf type'].value_counts()
plt.figure()
leaf_counts.plot(kind='bar')
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Distribution of the leaf type")
plt.xlabel("Type of the leaf")
plt.ylabel("species number")
plt.show()

fruit_counts = file['Fruit Type'].value_counts()
plt.figure()
fruit_counts.plot(kind='bar')
plt.xticks(rotation=45)
plt.tight_layout()
plt.title("Distribution of the fruit type")
plt.xlabel("Type of the fruit")
plt.ylabel("species number")
plt.show()
```
